```python
#df=web.DataReader('AAPL', start, end)
#df.to_csv('aapl.csv')
df=pd.read_csv('/Users/LM/Downloads/AAPL.csv', parse_dates=True, index_col=0)


# create empty list to store data
ThirdDayIncreased=[]
ThirDayDecreased=[]




```

```python
# Convert our Apple 'Close dataframe to list, this will help dealing with for loop

A = df['Close'].values.tolist()

print(A)

# using for loop, divide into 3 day periods, if 1st, 2nd  and 3rd day increased append it in ThirdDayIncreased=[]
# if 3rd day decreased append in ThirDayDecreased=[]

for i in range(len(A)-2):
    if A[i] < A[i+1] and A[i+1] < A[i+2]:
        ThirdDayIncreased.append(i)
    elif A[i] < A[i+1] and A[i+1] > A[i+2]:
        ThirDayDecreased.append(i)

print(ThirdDayIncreased)
print(ThirDayDecreased)

print(len(ThirdDayIncreased))
print(len(ThirDayDecreased))

PercentageIncreased=len(ThirdDayIncreased)/(len(ThirdDayIncreased)+len(ThirDayDecreased))

print(PercentageIncreased)


```
